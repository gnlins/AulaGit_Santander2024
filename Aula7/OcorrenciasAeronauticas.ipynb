![ModeloDados](https://raw.githubusercontent.com/gnlins/class_datasets/main/OcorrenciasAeronauticas/modelo_dados_cenipa.webp)

```python
MAIN_PATH = "https://raw.githubusercontent.com/gnlins/class_datasets/main/OcorrenciasAeronauticas"

ocorrencias = pd.read_csv(f"{MAIN_PATH}/ocorrencia.csv", sep=';',encoding='latin-1')

ocorrencia_tipo = pd.read_csv(f"{MAIN_PATH}/ocorrencia_tipo.csv", sep=';',encoding='latin-1')

recomendacao = pd.read_csv(f"{MAIN_PATH}/recomendacao.csv", sep=';',encoding='latin-1')

fator_contribuinte = pd.read_csv(f"{MAIN_PATH}/fator_contribuinte.csv", sep=';',encoding='latin-1')

aeronave = pd.read_csv(f"{MAIN_PATH}/aeronave.csv", sep=';',encoding='latin-1')
```

```python
ocorrencias.head()
```

## Associando Tabelas

```python
ocorrencias = ocorrencias.merge(ocorrencia_tipo,on='codigo_ocorrencia1', how='left' )
```

```python
ocorrencias = ocorrencias.merge(aeronave,on='codigo_ocorrencia2', how='left' )
```

```python
ocorrencias = ocorrencias.merge(fator_contribuinte,on='codigo_ocorrencia3', how='left' )
```

```python
ocorrencias = ocorrencias.merge(recomendacao,on='codigo_ocorrencia4', how='left' ) 
```
